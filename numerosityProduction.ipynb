```python
import torch
from diffusers import StableDiffusionPipeline, DPMSolverMultistepScheduler
import os
```

```python
def looper(object_ls, image_path = "D:/sd/"):

  model_id = "stabilityai/stable-diffusion-2-1"

  pipe = StableDiffusionPipeline.from_pretrained(model_id, torch_dtype=torch.float16)
  pipe.scheduler = DPMSolverMultistepScheduler.from_config(pipe.scheduler.config)
  pipe = pipe.to("cuda")

  for object in object_ls:
    obdir=image_path+object
    if not os.path.exists(obdir):
      os.makedirs(obdir)

    for numerosity in range(2,11):               # the target numerosity, default from 2 to 10. numerosity 1 was tested by mannully modifying this part.
      numdir=obdir+'/'+'/'+str(numerosity)           # generating the directory given the numerosity and category
      if not os.path.exists(numdir):
        os.makedirs(numdir)
      for a in range(50):             # the number of trials of each numerosity, default=50
        prompt=f"{numerosity} {object}"        # prompt generation
        image=pipe(prompt=prompt).images[0]     # image generation
        image.save(numdir + '/' +prompt+" "+str(a)+".png") # save image
```

```python
objects=['apples', 'butterflies', 'dots', 'people'] # for singular words, this list will be mannuly modified.
```

```python
looper(objects)
```
